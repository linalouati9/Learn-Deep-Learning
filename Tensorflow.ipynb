# MSE VS Cross-entropy ou Log Loss

```python
def mse(target, prediction):
  loss = (prediction - target) ** 2
  return loss
```

```python
def log_loss(target, prediction):
  loss = -1 * np.log(prediction)
  return loss
```

```python
prediction = np.arange(0, 1, 0.01)
prediction
```

```python
target = np.ones_like(prediction)
target
```

```python
loss_mse = mse(target, prediction)
loss_ce = log_loss(target, prediction)
```

```python
# print(loss_mse); print(loss_mse)
```

```python
plt.plot(prediction, loss_mse, c='red', label='MSE')
plt.plot(prediction, loss_ce, c='blue', label='Cross entropy')
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.show()
```

```python
def log_loss(target, prediction):
  loss = -1 * np.log(1 - prediction)
  return loss
```

```python
target = np.zeros_like(prediction)
target
```

```python
loss_mse = mse(target, prediction)
loss_ce = log_loss(target, prediction)
```

```python
plt.plot(prediction, loss_mse, c='red', label='MSE')
plt.plot(prediction, loss_ce, c='blue', label='Cross entropy')
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.show()
```

```python
def plot_learning_curve(history):
  plt.plot(list(range(len(history['loss']))), history['loss'], label='loss')
  plt.plot(list(range(len(history['val_loss']))), history['val_loss'], label='val_loss')
  plt.xlabel('Epochs')
  plt.ylabel('Loss')
  plt.title('Loss Curve')
  plt.legend(loc='best')
  plt.show()
```

```python
def plot_accuracy_curve(history):
  plt.plot(list(range(len(history['accuracy']))), history['accuracy'], label='accuracy')
  plt.plot(list(range(len(history['val_accuracy']))), history['val_accuracy'], label='val_accuracy')
  plt.xlabel('Epochs')
  plt.ylabel('Accuracy')
  plt.title('Accuracy Curve')
  plt.legend(loc='best')
  plt.show()
```

```python
plot_learning_curve(h.history)
plot_accuracy_curve(h.history)
```

```python
# Overfitting
# Un model tellement entrain√© il commence √† apprendre betement
# Lorsqu'on a augmenter le nombre de hidden layers de notre mod√©le
# plus de 100K params a √©t√© utilis√© pour une augmentation de accuracy de 2% => Inutile
```

# Combattre le overfitting : Dropout

# Adam Optimizer : Le concept de Momentum

```python
model = Sequential()
model.add(Dense(units=784, activation="relu"))
model.add(tf.keras.layers.Dropout(0.5))
model.add(Dense(units=128, activation="relu"))
model.add(tf.keras.layers.Dropout(0.3))
model.add(Dense(units=10, activation="softmax"))

adam = tf.keras.optimizers.Adam(
    learning_rate=0.01,
    beta_1=0.9,
    beta_2=0.999
)
model.compile(loss="categorical_crossentropy", optimizer=adam, metrics = ['accuracy'])
h = model.fit(flatten_training_images, training_labels, epochs=20, validation_data=(flatten_test_images, test_labels))
```

# Sauvegarder son Meilleur Mod√®le - Le Callback ModelCheckpoint


```python
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
best_model = 'best_model.h5'

model_ckp = ModelCheckpoint(filepath=best_model,
                            #save_weights_only=True,
                            monitor="val_accuracy",
                            mode="max",
                            save_best_only=True)
```

```python
model = Sequential()
model.add(Dense(units=784, activation="relu"))
model.add(tf.keras.layers.Dropout(0.5))
model.add(Dense(units=128, activation="relu"))
model.add(tf.keras.layers.Dropout(0.3))
model.add(Dense(units=10, activation="softmax"))

model.compile(loss="categorical_crossentropy", optimizer=SGD(learning_rate=0.01), metrics = ['accuracy'])
h = model.fit(flatten_training_images, training_labels,
              epochs=3,
              validation_data=(flatten_test_images, test_labels),
              callbacks=[model_ckp])
```

```python
model_save = tf.keras.models.load_model('best_model.h5')
model_save.evaluate(flatten_test_images, test_labels)
```

# Le Callback Early Stopping


```python
# les callbacks sont utiliser pour plusieurs raisons exemple :
# changer le learning rate en runtime de l entrainement du model,
# D√©s qu'on atteind une certaine accuracy, on envoie un message whats'up, email,
# Live visualisation ..
```

```python
# Le fait de pouvoir arr√©ter le train tot
stop = tf.keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=2)
```

```python
model = Sequential()
model.add(Dense(units=784, activation="relu"))
model.add(tf.keras.layers.Dropout(0.5))
model.add(Dense(units=128, activation="relu"))
model.add(tf.keras.layers.Dropout(0.3))
model.add(Dense(units=10, activation="softmax"))

model.compile(loss="categorical_crossentropy", optimizer=SGD(learning_rate=0.01), metrics = ['accuracy'])
h = model.fit(flatten_training_images, training_labels,
              epochs=1000,
              validation_data=(flatten_test_images, test_labels),
              callbacks=[model_ckp, stop])
```
